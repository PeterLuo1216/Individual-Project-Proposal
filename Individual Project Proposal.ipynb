First Analysis: Self-Identity and Mental Health

Research Question: Is there a link between self-identity and mental health scores in the CSCS dataset?

Variables:
1:DEMO_identity_lgbtq: LGBTQ+ self-identification
2:PSYCH_body_self_image_questionnaire_negative_affect_score: Negative emotions about body image (higher scores = more negativity)
3:PSYCH_big_five_inventory_neurotic_score: Emotional instability (higher scores = more instability)

Visualizations:
Violin Plots: Show body dissatisfaction and negative affect scores for LGBTQ+ vs. non-LGBTQ+ groups, revealing density and variation.
Grouped Bar Charts: Compare mean neuroticism between LGBTQ+ and non-LGBTQ+ groups.

Method: Bootstrapping
Separate data by identity, analyze body dissatisfaction and neuroticism.
Draw 1,000 bootstrap samples, calculating the mean or median for each group.
Generate confidence intervals and compare them. Non-overlapping intervals suggest meaningful differences.

Hypothesis: LGBTQ+ individuals may report higher dissatisfaction and neuroticism scores than non-LGBTQ+ individuals.

Relevance: Results will highlight mental health disparities, supporting awareness for underrepresented groups.

Second Analysis: Age and Mental Health

Research Question: How does age affect mental health scores, like neuroticism and body dissatisfaction?

Variables:
1:DEMO_age: Age
2:PSYCH_big_five_inventory_neurotic_score: Emotional instability
3:PSYCH_big_five_inventory_extraverted_score: Extraversion
4:PSYCH_body_self_image_questionnaire_height_dissatisfaction_score: Height dissatisfaction

Visualizations:
Box Plots: Display neuroticism and body dissatisfaction across age groups.
Summary Statistics: Present mean and standard deviation for each age group.

Method: Simple Linear Regression
Define age as the independent variable, mental health scores as dependent.
Fit a model to determine how scores change with age.

Hypothesis: Older individuals may have lower neuroticism, while extraversion may change unpredictably with age.

Relevance: Identifying age-related mental health patterns can help in targeted interventions.
